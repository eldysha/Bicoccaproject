```python
import pandas as pd
import numpy as np
# disable chained assignments
pd.options.mode.chained_assignment = None 
# dataset import
df_store = pd.read_csv('https://raw.githubusercontent.com/gdv/foundationsCS-2018/master/ex-data/project/googleplaystore.csv')
df_reviews = pd.read_csv('https://raw.githubusercontent.com/gdv/foundationsCS-2018/master/ex-data/project/googleplaystore_user_reviews.csv')
df_store.head()
```

#    1.Convert the app sizes to a number
   

```python
#remove invalid record
df_store=df_store.drop(df_store.index[10472])
df_store= df_store.reset_index(drop=True)
```

```python
# convert M and k size
i=0
for n in df_store["Size"]:
    if str(df_store["Size"][i]).endswith("M"):
        df_store["Size"][i]=df_store["Size"][i][:-1]
        df_store["Size"][i]=float(df_store["Size"][i]) 
        df_store["Size"][i]=df_store["Size"][i]*10**6
    if str(df_store["Size"][i]).endswith("k"):
        df_store["Size"][i]=df_store["Size"][i][:-1]
        df_store["Size"][i]=float(df_store["Size"][i]) 
        df_store["Size"][i]=df_store["Size"][i]*10**3
    i=i+1
df_store.head()
```

#    2.Convert the number of installs to a number

```python
# remove unwanted characters and transform values into number
i=0
for k in df_store["Installs"]:
    df_store["Installs"][i]=df_store["Installs"][i].replace(",","")
    df_store["Installs"][i]=df_store["Installs"][i].replace("+","")
    df_store["Installs"][i]=int(df_store["Installs"][i])
    i=i+1
df_store.head()
```

# 3.Transform “Varies with device” into a missing value

```python
df_store=df_store.replace('Varies with device',np.nan)
df_store.head()
```

# 4.Convert Current Ver and Android Ver into a dotted number (e.g. 4.0.3 or 4.2)

```python
import re
#define a function to remove all character except for numbers and point character
def convert_number(var):
    try:
        return re.sub("[^0-9.]","",var)
               
    except:
        return np.nan
```

```python
# apply previous function
df_store["Current Ver"]=df_store.apply(lambda x:convert_number(x["Current Ver"]),axis=1)
df_store["Android Ver"]=df_store.apply(lambda x:convert_number(x["Android Ver"]),axis=1)
df_store.head()
```

# 5. Remove the duplicates

```python
# Keep only first value for the duplicates records recognized through name and size
df_store = df_store.drop_duplicates(subset=['App','Size'], keep='first', inplace=False)
df_store= df_store.reset_index(drop=True)
df_store.head()
```

# 6. For each category, compute the number of apps

```python
# Count the Apps for each category
df_store.groupby(['Category']).size().reset_index(name='counts')
```

# 7. For each category, compute the average rating

```python
# Average for category
df_store.groupby(df_store['Category']).mean()['Rating']
```

# 8. Create two dataframes: one for the genres and one bridging apps and genres. So that, for instance, the app Pixel Draw - Number Art Coloring Book appears twice in the bridging table, once for Art & Design, once for Creativity

```python
#8.1 create dataframe for genres
# Create a dataframe with splitted genres
d_genres=pd.DataFrame(df_store['Genres'])
d_genres=d_genres['Genres'].str.split(';', 1, expand=True)
d_genres.head()
```

```python
#create two dataframe for the corrisponding splitted genres column
d_genres0=pd.DataFrame(d_genres[0]) #column genres 1
d_genres1=pd.DataFrame(d_genres[1]) #column genres 2

#rename the column of the two dataframe with a unique name
d_genres1.columns=['Genre']
d_genres1=d_genres1.dropna() #remove NA value
d_genres0.columns=['Genre']

#Concatenate 
d_genre_conc=pd.concat([d_genres1, d_genres0], ignore_index=True, sort=False)
d_genre_conc= d_genre_conc.reset_index(drop=True)
d_genre_conc = d_genre_conc.groupby(d_genre_conc['Genre']).sum()
d_genre_conc
```

```python
#8.2 create bridge dataframe
#create two dataframe with the splitted column "genres"
d_bridge=pd.DataFrame(df_store[['App','Genres']])
d_bridge[['Genre1','Genre2']]=d_bridge['Genres'].str.split(';', 1, expand=True)
d_bridge=pd.DataFrame(d_bridge[['App','Genre1','Genre2']])
d_bridge.head()
```

```python
# create a unique dataframe concatenating the 1st dataframe and the 2nd dataframe in a sequential way
i=len(d_bridge)
i=i+1
j=0
#add a new row with the value of the second dataframe
for n in d_bridge['App']:
    d_bridge.loc[i]=[d_bridge['App'][j],d_bridge['Genre2'][j],d_bridge['Genre2'][j]]
    i=i+1
    j=j+1
d_bridge=pd.DataFrame(d_bridge[['App','Genre1']])
d_bridge=d_bridge.dropna()
d_bridge.columns=['App','Genre'] #rename columns
d_bridge= d_bridge.reset_index(drop=True)
d_bridge.head()
```

# 11. For each app, compute the approximate income, obtain as a product of number of installs and price.

```python
#apply the function which mantains only numbers and point character
df_store["Price"]=df_store.apply(lambda x:convert_number(x["Price"]),axis=1)
```

```python
# Calculate the approximate income as numbers of installation*price
i=0
df_store["Income"]=None
for n in df_store['App']:
    df_store["Income"][i] =int(df_store["Installs"][i])*float(df_store['Price'][i])
    i=i+1
    
df_store["Income"]=df_store["Income"].astype(float)
df_store.sort_values(by=['Income'],ascending=False)[:20]
```

# 12. For each app, compute its minimum and maximum Sentiment_polarity

```python
#compute minimum, compute maximum
max_test=df_reviews['Sentiment_Polarity'].groupby(df_reviews['App']).max().reset_index(name='max_sent')
min_test=df_reviews['Sentiment_Polarity'].groupby(df_reviews['App']).min().reset_index(name='min_sent')
#merge togheter the two results
result = pd.merge(min_test, max_test, left_on = 'App', right_on = 'App', how = 'inner')
result[:20]
```
